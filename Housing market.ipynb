```python
train = pd.read_csv("train.csv")
train.head()
```

```python
threshold = 1000
train = train.drop(list(train.isna().sum()[train.isna().sum() > threshold].index), axis=1)
```

```python
X = train.drop(['SalePrice'], axis=1)
y = train['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
train_data = X_train.copy()
train_data['SalePrice'] = y_train
train_data
```

```python
cor_threshold = train_data.corr()['SalePrice'].median() - 0.1
features = list(train_data.corr()[np.abs(train_data.corr()['SalePrice']) > cor_threshold]['SalePrice'].index)
features.remove('SalePrice')

features
```

```python
X_train = X_train[features]
X_test = X_test[features]
```

```python
X_train = X_train.dropna()
X_test = X_test.dropna()
```

```python
def drop_IDS(X, y):
    drop = []
    for y_id in y.index:
        if y_id not in X.index:
            drop.append(y_id)
    return drop
```

```python
y_train = y_train.drop(drop_IDS(X_train, y_train))
y_test = y_test.drop(drop_IDS(X_test, y_test))
```

```python
pipeline1 = Pipeline([
    ('scalar', StandardScaler()),
    ('linear', LinearRegression(positive=True))
])
pipeline1.fit(X_train, y_train)
```

```python
pipeline1.score(X_test, y_test)
```
